# Data Tidying And Cleaning

Data Science процеса е OSEMN. - Obtain, Scrub, Explore, Model, iNtepret. 
iNtepret - направлява целия процес.

Имаме таблици които са двумерни. Имаме редове и колони. Всеки един ред е наблюдение.
Неслучайно наричаме Data set - редовете трябва да бъдат уникални. 
Един ред отговаря за един цял запис.
Релационна алгебра - говорим за 4-та нормална форма. 
Нека Имаме матрица от числа. Ако я разглеждаме като трансформация, всеки базисен вектор е написан като една колона. Защо колони - по конвенция. За да може да умножаваме някакви данни като ред по стълба.
В случая една колона е едно измерение на данните, една променлива.
Един запис представлява линейна комбинация от всеки един базисен вектор, умножен по коефициента, който в вътре. Един запис представлява една точка в това многомерно пространство. Така работи ML, Pandas.
Една клетка съдържа 1 запис за 1 променлива. В клетките трябва да има скаларни величини. - какво е скалар - въпроса е да бъде 1 нещо. Ако има речник или сет - не е допустимо.
1 таблица, dataset в идеалния случай представлява матрица от данните и форматът, в който са данните, трябва да спазват тези изисквания.
Структурирани данни - такива, които са в таблица. Неструктурирани - някакви други.
Ако искаме да добавяме нов запис, добавяме нов ред. Но ако трябва за това да редактираме съществуващия - проблем.

```python
pew = pd.read_csv("data/pew.csv")
```

```python
pew
```

```python
Това е pivot table. 
Първо което трябва да разберем е кои ни са променливите:
Religion, Salary range, count.
Правим операцията, обратна на създаването на Pivot Table, и тя се нарича melt.
```

```python
pew.melt(id_vars = ["religion"])
```

Това оставя religion и разделя данните на отделни редове в съответната колона.

```python
pew_tidy = pew.melt(id_vars = ["religion"], var_name = "income", value_name = "frequency")
```

```python
pew_tidy
```

```python
pew.columns, pew_tidy.columns
```

Ако имаме категорийни променливи, можем да укажем, че това е така. Това пести и памет.

```python
pew_tidy.religion = pew_tidy.religion.astype("category")
pew_tidy.income = pew_tidy.income.astype("category")
```

```python
pew_tidy.dtypes
```

```python
pew_tidy.memory_usage()
```

Тази оптимизация се прави при големи datasets и много дълги имена на категориите.
